```python
import tensorflow as tf
from keras_preprocessing.image import ImageDataGenerator
import numpy as np
```

```python
training_datagen = ImageDataGenerator(
    rescale = 1./255,
	fill_mode ='nearest'
)
```

```python
validation_datagen = ImageDataGenerator(
    rescale = 1./255
)
```

```python
train_generator = training_datagen.flow_from_directory(
	'drive/MyDrive/Datasets/horses_or_humans/training',
	target_size = (150,150),
	class_mode = 'categorical',
    batch_size = 8
)
```

```python
validation_generator = validation_datagen.flow_from_directory(
	'drive/MyDrive/Datasets/horses_or_humans/test',
	target_size = (150,150),
	class_mode = 'categorical',
    batch_size = 1
)
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(64, (3,3), activation='relu', input_shape=(150, 150, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(2, activation='softmax')
])
```

```python
model.summary()
```

```python
model.compile(loss = 'categorical_crossentropy', optimizer = 'rmsprop', metrics = ['accuracy'])
```

```python
model.fit(train_generator, epochs = 25, steps_per_epoch = 5, verbose = 1)
```

```python
validation_results = []
for validation_file, test_label in validation_generator:
    img_data = validation_file
    if np.argmax(test_label[0]) == np.argmax(model.predict(img_data)[0]):
        print('True')
        validation_results.append(True)
    else:
        print('False')
        validation_results.append(False)
```
