# Simple Computer Vision Example With Fashion MNIST Dataset

We use the fashion mnist dataset which contains 70,000 images of different objects and train a simple model to recognize them.

```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

## Let's Load And Prepare The Data

```python
mnist = tf.keras.datasets.fashion_mnist
```

```python
(training_images, training_labels), (test_images, test_labels) = mnist.load_data()
```

```python
img_index = 0
plt.imshow(training_images[img_index]), training_labels[img_index]
```

### Normalize The Images

```python
tf.reduce_max(training_images) # max value in the image
```

```python
tf.reduce_min(training_images) # min value in the image
```

```python
training_images  = training_images / 255.0
test_images = test_images / 255.0
```

## Let's Build The Model

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(1024, activation=tf.nn.relu),
    tf.keras.layers.Dense(10, activation=tf.nn.softmax)
])
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

## Let's Fit The Model

```python
model.fit(training_images, training_labels, epochs=20)
```

```python
model.evaluate(test_images, test_labels)
```
