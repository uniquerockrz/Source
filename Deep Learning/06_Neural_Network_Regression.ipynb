# Regression In Neural Networks With Tensor Flow

We will build a regression model with Tensorflow using Neural Networks. We will be predicting a number based on other numbers. 

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
print(tf.__version__)
```

## Creating Some Data

```python
# create the features
X = np.array([-7.0, -4.0, -1.0, 2.0, 5.0, 8.0, 11.0, 14.0])


# create the labels
y = np.array([3.0, 6.0, 9.0, 12.0, 15.0, 18.0, 21.0, 24.0])
```

```python
# visualize
plt.scatter(X, y)
```

## Input & Output Shapes

```python
# Lets look at the shapes of the dataset by default
X.shape, y.shape
```

```python
# Turn numpy arrays into tensors
X = tf.constant(X),
y = tf.constant(y),

X, y
```

## Modelling In Tensorflow

### Create The Model

```python
tf.random.set_seed(42)
```

```python
# Create the model using Sequential API

model = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])
```

### Compile The Model

```python
model.compile(loss=tf.keras.losses.mae, optimizer=tf.keras.optimizers.SGD(), metrics=['mae'])
```

### Fit The Model

```python
model.fit(tf.expand_dims(X, axis=-1), tf.expand_dims(y, axis=-1), epochs=5)
```

```python
# another way of creating the model
model2 = tf.keras.Sequential()
model2.add(tf.keras.layers.Dense(1))
```

### Make A Prediction

```python
model.predict([[17.0]])
```

## Improving The Model

```python
# add more epocs
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])

model.compile(loss=tf.keras.losses.mae, optimizer=tf.keras.optimizers.SGD(), metrics=['mae'])

model.fit(tf.expand_dims(X, axis=-1), tf.expand_dims(y, axis=-1), epochs=100)
```

```python
model.predict([[17.0]])
```

```python
# add one more neurons and another layer
model = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(1)
])

model.compile(loss=tf.keras.losses.mae, optimizer=tf.keras.optimizers.SGD(), metrics=['mae'])

model.fit(tf.expand_dims(X, axis=-1), tf.expand_dims(y, axis=-1), epochs=100)
```

```python
model.predict([[17.0]])
```

```python
# add more layers
model = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(1)
])

model.compile(loss=tf.keras.losses.mae, optimizer=tf.keras.optimizers.SGD(), metrics=['mae'])

model.fit(tf.expand_dims(X, axis=-1), tf.expand_dims(y, axis=-1), epochs=100)
```

```python
model.predict([[17.0]])
```

```python
# add Adam optimizer and increase the epochs
model = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(1)
])

model.compile(loss=tf.keras.losses.mae, optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001), metrics=['mae'])

model.fit(tf.expand_dims(X, axis=-1), tf.expand_dims(y, axis=-1), epochs=200)
```

```python
# change the activation function
model = tf.keras.Sequential([
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(1)
])

model.compile(loss=tf.keras.losses.mae, optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001), metrics=['mae'])

model.fit(tf.expand_dims(X, axis=-1), tf.expand_dims(y, axis=-1), epochs=200)
```

```python
model.predict([[17.0]])
```
