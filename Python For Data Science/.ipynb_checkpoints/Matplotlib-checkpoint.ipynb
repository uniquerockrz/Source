# Matplotlib Examples

Some of the widely used matplotlib examples

## Bootstrap The Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Choose Style

```python
# plt.style.use('ggplot')
plt.style.use('fivethirtyeight')
```

```python
# set seed
np.random.seed(42)
```

```python
# lets generate some random integers to plot
X = np.random.randint(0, 100, 100)
Y = np.random.randint(0, 100, 100)
X_normal = np.random.normal(0.0, 50.0, 100)
Y_normal = np.random.normal(0.0, 50.0, 100)
X_with_space = np.linspace(0, 100, 100)
Y_with_space = np.linspace(0, 100, 100)
```

## 1D Plots

### Line Chart

```python
# create a line chart with only one data point
fig_1d_line1, ax_1d_line1 = plt.subplots(1, 1)
ax_1d_line1.plot(X, color='red')
```

```python
# create a line chart with multiple data points
fig_1d_line2, ax_1d_line2 = plt.subplots(1, 1)
ax_1d_line2.plot(X, color='blue')
ax_1d_line2.plot(Y, color='green')
ax_1d_line2.plot(X_normal, color='red')
```

### Pie Chart

```python
fig_pie1, ax_pie1 = plt.subplots(1, 1, figsize=(3, 3))
ax_pie1.pie([1, 2, 3, 4])
```

### Histograms

```python
# figure with multiple axes
fig_1d_hist1, ax_1d_hist1 = plt.subplots(2, 2, figsize=(10, 10))
# single histogram plot
ax_1d_hist1[0, 0].hist(X, bins=20)
# multi histogram plot
ax_1d_hist1[0, 1].hist(X_normal, bins=20, color='red', label='X Normal')
ax_1d_hist1[0, 1].hist(Y_normal, bins=20, color='blue', label='Y Normal')
ax_1d_hist1[0, 1].legend()
# density plots
ax_1d_hist1[1, 0].hist(X_normal, bins=20, color='red', label='X Normal', density=True)
ax_1d_hist1[1, 0].hist(Y_normal, bins=20, color='blue', label='Y Normal', density=True)
ax_1d_hist1[1, 0].legend()
# with steps
ax_1d_hist1[1, 1].hist(X_normal, bins=20, color='red', label='X Normal', histtype='step')
ax_1d_hist1[1, 1].hist(X_normal, bins=20, color='blue', label='X Normal')
```

## 2d Plots

### Simple Line Charts

```python
fig_2d_line, ax_2d_line = plt.subplots(1, 1)
# plot multiple line plots
ax_2d_line.plot(X_with_space[:20], Y[:20], label = 'X over Y till 20', marker='o', linewidth=3)
ax_2d_line.plot(X_with_space[:20], Y[20:40], label = 'X over Y till 40', marker='*', linewidth=3)
ax_2d_line.set_xlabel('X', color='gold')
ax_2d_line.set_ylabel('Y', color='red')
ax_2d_line.set_title('X over Y', fontsize=18)
ax_2d_line.legend()
```

### Scatter Plots

```python
fig_scatter1, ax_scatter_1 = plt.subplots(1, 2, figsize=(8,4))
ax_scatter_1[0].scatter(X, Y)
# with alpha
ax_scatter_1[1].scatter(X, Y, alpha=0.2)
```

### Bar Charts

```python
# simple bar charts
fig_bar1, ax_bar1 = plt.subplots(1, 1)
ax_bar1.bar(X, Y)
```

```python
# stacked bar charts
fig_bar2, ax_bar2 = plt.subplots(1, 1, figsize=(8, 8))
ax_bar2.bar(X, Y, label='Y')
ax_bar2.bar(X, Y_normal, bottom=Y, label='Y_normal')
ax_bar2.bar(X, Y_with_space, bottom=Y_normal, label='Y_with_space')
ax_bar2.legend()
```

### Stack plots

```python
# stack plots
fig_stack1, ax_stack1 = plt.subplots(1, 1, figsize=(4, 4))
X_stack = np.arange(0, 20, 2)
ax_stack1.stackplot(X_stack, Y[:10])
ax_stack1.stackplot(X_stack, Y[10:20])
ax_stack1.stackplot(X_stack, Y[20:30])
```

```python
# stack plots with v stack
fig_stack2, ax_stack2 = plt.subplots(1, 1, figsize=(4, 4))
X_stack = np.arange(0, 20, 2)
Y_stack = np.vstack([Y[:10], Y[10:20], Y[20:30]])
ax_stack2.stackplot(X_stack, Y_stack)
```

### Box Plots

```python
fig_box1, ax_box1 = plt.subplots(1, 1, figsize=(2, 2))
ax_box1.boxplot([X, Y, X_normal, Y_normal])
```

## 3D plots

### Contours

```python
fig_contour1, ax_contour1 = plt.subplots(1, 1)
X_contour = np.linspace(0, 100, 100, dtype=np.int32)
Y_contour = np.linspace(0, 100, 100, dtype=np.int32)
Z_contour = np.random.randint(0, 2, size=(100, 100))
ax_contour1.contourf(X_contour, Y_contour, Z_contour, levels=1, colors=['red', 'blue'])
```

### 3D Scatter Plots

```python
fig_3d_scatter1, ax_3d_scatter1 = plt.subplots(1, 1, subplot_kw={"projection": "3d"})
ax_3d_scatter1.scatter(X, Y, X_normal)
```
